```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib as mpl
import analytics
```

```python
data = analytics.get_data()
brn_data = data['BRN']
khm_data = data['KHM']
ldn_data = data['IDN']
lao_data = data['LAO']
mys_data = data['MYS']
mmr_data = data['MMR']
phl_data = data['PHL']
sgp_data = data['SGP']
tha_data = data['THA']
vnm_data = data['VNM']
```

```python
gdp_phl = phl_data[phl_data['indicator_code'] == 'NY.GDP.MKTP.KD']
pl = gdp_phl.plot(kind='line', x='year', y='value', title='PHL GDP in 2005 USD over time', legend=False, figsize=(8, 6))
pl.set_xlabel("Year")
pl.set_ylabel("GDP in 2005 USD")
```

```python
pop_dnst = phl_data[phl_data['indicator_code'] == 'EN.POP.DNST']
le = phl_data[phl_data['indicator_code'] == 'SP.DYN.LE00.IN']
fig, ax = mpl.pyplot.subplots()
pl = pd.concat([le]).groupby('indicator_code').plot(kind='line', x='year', y='value', title='Life expectancy in years over time', ax=ax, legend=False, figsize=(8, 6))
ax.set_xlabel('Year')
ax.set_ylabel('Life expectancy in years')

fig, ax = mpl.pyplot.subplots()
pl = pd.concat([pop_dnst]).groupby('indicator_code').plot(kind='line', x='year', y='value', title='Population density over time', ax=ax, legend=False, figsize=(8, 6))
ax.set_xlabel('Year')
ax.set_ylabel('Population density (p/km^2) ')
```

```python
low_ten = phl_data[phl_data['indicator_code'] == 'SI.DST.FRST.10']
low_twenty = phl_data[phl_data['indicator_code'] == 'SI.DST.05TH.20']
high_twenty = phl_data[phl_data['indicator_code'] == 'SI.DST.FRST.20']
high_ten = phl_data[phl_data['indicator_code'] == 'SI.DST.10TH.10']
fig, ax = mpl.pyplot.subplots()
pl = pd.concat([high_ten, high_twenty, low_twenty , low_ten]).groupby('indicator_code').plot(kind='line', x='year', y='value', title='Income distribution across quartiles over time', ax=ax, legend=True, figsize=(8, 6))
ax.set_xlabel('Year')
ax.set_ylabel('Percent of national income')
ax.legend([ax.get_lines()[0], ax.get_lines()[1], ax.get_lines()[2], ax.get_lines()[3]], ['Highest ten', 'Highest twenty', 'Lowest twenty', 'Lowest ten'])
```

```python
urban_pop_p = phl_data[phl_data['indicator_code'] == 'SP.URB.TOTL.IN.ZS']
fig, ax = mpl.pyplot.subplots()
pl = pd.concat([urban_pop_p]).groupby('indicator_code').plot(kind='line', x='year', y='value', title='Urban population percentage over time', ax=ax, legend=False, figsize=(8, 6))
ax.set_xlabel('Year')
ax.set_ylabel('Urban Population Percentage')
```

```python
trade = phl_data[phl_data['indicator_code'] == 'NE.TRD.GNFS.ZS']
fig, ax = fig, ax = mpl.pyplot.subplots()
pl = pd.concat([urban_pop_p, trade]).groupby('indicator_code').plot(kind='line', x='year', y='value', title='Urban Population and Trade over time', ax=ax, legend=True, figsize=(8, 6))
ax.set_xlabel('Year')
ax.set_ylabel('Percentage')
ax.legend([ax.get_lines()[0], ax.get_lines()[1]], ['Urban population as percentage of total population', 'Trade as percentage of GDP'])
```

```python
pop_grow = phl_data[phl_data['indicator_code'] == 'SP.POP.GROW']
fig, ax = fig, ax = mpl.pyplot.subplots()
pl = pd.concat([pop_grow]).groupby('indicator_code').plot(kind='line', x='year', y='value', title='Population growth over time', ax=ax, legend=True, figsize=(8, 6))
ax.set_xlabel('Year')
ax.set_ylabel('Percentage of Population Growth')
ax.legend([ax.get_lines()[0]], ['Population Growth'])
```

```python
sec_prog = phl_data[phl_data['indicator_code'] == 'SE.SEC.PROG.ZS']
prm_xpd = phl_data[phl_data['indicator_code'] == 'SE.XPD.MPRM.ZS']
fig, ax = fig, ax = mpl.pyplot.subplots()
pl = pd.concat([prm_xpd, sec_prog]).groupby('indicator_code').plot(kind='line', x='year', y='value', title='Primary school expenditure and secondary education progression over time', ax=ax, legend=True, figsize=(8, 6))
ax.set_xlabel('Year')
ax.set_ylabel('Percentage')
ax.legend([ax.get_lines()[0], ax.get_lines()[1]], ['Primary school expenditure as percent of government expenditure', 'Percent of students that reach secondary education'])
```

```python
pov_gap = phl_data[phl_data['indicator_code'] == 'SI.POV.GAPS']
fig, ax = fig, ax = mpl.pyplot.subplots()
pl = pd.concat([pov_gap, sec_prog]).groupby('indicator_code').plot(kind='line', x='year', y='value', title='Secondary school progression and poverty gap over time', ax=ax, legend=True, figsize=(8, 6))
ax.set_xlabel('Year')
ax.set_ylabel('Percentage')
ax.legend([ax.get_lines()[0], ax.get_lines()[1]], ['Secondary school progression', 'Poverty Gap'])
```
